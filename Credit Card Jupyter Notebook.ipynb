```python
# Task 1

import pandas as pd

# header to None as it doesn't have column names
cc_apps = pd.read_csv(r"C:\Users\44741\Downloads\cc_approvals.data", header= None)

print(cc_apps.head())
```

```python
# Task 2

print("Summary Statistics:")
print(cc_apps.describe())

print("\nDataset Information:")
print(cc_apps.info())

print("\nLast 17 Rows of the Dataset:")
print(cc_apps.tail(17))

```

```python
#Task 3

from sklearn.model_selection import train_test_split

cc_apps = cc_apps.drop(columns=[11, 13])

# Split into train and test sets
cc_apps_train, cc_apps_test = train_test_split(cc_apps, test_size=0.33, random_state=42)

print("Training set shape:", cc_apps_train.shape)
print("Test set shape:", cc_apps_test.shape)
```

```python
# Task 4

import numpy as np

# Replace the question marks with NaN in the train and test sets
cc_apps_train = cc_apps_train.replace('?', np.nan)
cc_apps_test = cc_apps_test.replace('?', np.nan)
```

```python
# Task 5

# Select only numeric columns for mean imputation
numeric_cols = cc_apps_train.select_dtypes(include=['number']).columns

# Impute missing values in numeric columns using mean (using inplace=True)
cc_apps_train[numeric_cols].fillna(cc_apps_train[numeric_cols].mean(), inplace=True)
cc_apps_test[numeric_cols].fillna(cc_apps_train[numeric_cols].mean(), inplace=True)  

# Count the number of NaNs in the datasets 
print("Missing values after mean imputation:")
print(cc_apps_train.isnull().sum())
print(cc_apps_test.isnull().sum())

```

```python
# Task 6

# Iterate over each column of cc_apps_train
for col in cc_apps_train.columns:
    # Check if the column is of object type 
    if cc_apps_train[col].dtype == 'object':
        # Impute missing values with the most frequent value 
        mode_value = cc_apps_train[col].mode()[0]  
        
        cc_apps_train[col] = cc_apps_train[col].fillna(mode_value) #Replaces all missing values in the training set with the most frequent value
        cc_apps_test[col] = cc_apps_test[col].fillna(mode_value)  
    

# Count the number of NaNs in the dataset and print the counts to verify
print("Missing values after mode imputation (categorical columns):")
print(cc_apps_train.isnull().sum())
print(cc_apps_test.isnull().sum())
```

```python
# Task 7:

# Convert the categorical features in the train and test sets independently
cc_apps_train = pd.get_dummies(cc_apps_train) # Creates dummy variables for each categorical column.
cc_apps_test = pd.get_dummies(cc_apps_test)

# Reindex the columns of the test set aligning with the train set
cc_apps_test = cc_apps_test.reindex(columns=cc_apps_train.columns, fill_value=0) 
# ensures test set has the same columns as training and if there are missing values it is filled with 0


print("First 5 rows of training set after encoding:\n", cc_apps_train.head())
print("\nFirst 5 rows of test set after encoding:\n", cc_apps_test.head())
```

```python
# Task 8: 

# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Segregate features and labels into separate variables
X_train, y_train = cc_apps_train.iloc[:, :-1].values, cc_apps_train.iloc[:, -1].values
X_test, y_test = cc_apps_test.iloc[:, :-1].values, cc_apps_test.iloc[:, -1].values

# rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1)) # scales data to a range of [0,1]
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)


print("First 5 rows of rescaled training set:\n", rescaledX_train[:5])
print("\nFirst 5 rows of rescaled test set:\n", rescaledX_test[:5])
```

```python
# Task 9

# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Instantiate a LogisticRegression classifier with default parameter values
logreg = LogisticRegression() # max_iter=100, C=1.0 (inverse of regularization strength), solver='lbfgs' (optimization algorithm for small datasets).

# Fit logreg to the training set
logreg.fit(rescaledX_train, y_train)
```

```python
# Task 10

# Import confusion_matrix
from sklearn.metrics import accuracy_score, confusion_matrix

# Use logreg to predict instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

# # Calculates the accuracy score by comparing actual labels with the predicted values .
print("Accuracy of logistic regression classifier:", accuracy_score(y_test, y_pred))
# Calculates the accuracy score by comparing actual labels with the predicted values .

# Print the confusion matrix of the logreg model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Task 11

# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the grid of values for tol and max_iter
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200]

# Create a dictionary where tol and max_iter are keys and their values are lists of possible values
param_grid = dict(tol=tol, max_iter=max_iter)
```

```python
# Task 12

# Instantiate GridSearchCV with the required parameters
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5) # performs 5 cross validations

# Fit grid_model to the data
grid_model_result = grid_model.fit(rescaledX_train, y_train) # Stores the best-performing model in grid_model_result

# Summarize results
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))

# best_score correlates to the highest cross-validation accuracy achieved
# best_params correlates to the best tol and max_iter values 



# Extract the best model and evaluate it on the test set
best_model = grid_model_result.best_estimator_
print("Accuracy of logistic regression classifier:", best_model.score(rescaledX_test, y_test))
```
